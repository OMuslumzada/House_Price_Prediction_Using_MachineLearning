#  House Price Prediction using Machine Learning

##  Project Overview
This project aims to predict house prices using machine learning techniques.  
The dataset contains various features such as lot area, number of rooms, and overall quality.

We build a model, evaluate its performance, and analyze the results.



## 1. Data Integration and Model Training
This section handles the environment setup, data validation, and the training of our Random Forest model.

```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Configuration and Data Verification
DATA_PATH = 'data/train.csv'
FEATURES = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
TARGET = 'SalePrice'

def load_and_train():
    # check if the file exists and has content
    if os.path.exists(DATA_PATH) and os.path.getsize(DATA_PATH) > 0:
        df = pd.read_csv(DATA_PATH)
        print(f"Success: Loaded {len(df)} rows from {DATA_PATH}")
    else:
        print("Warning: local data not found. Initializing 'Simulation Mode' with synthetic data for testing.")
        # Creating synthetic data so you can still run the code to check for logic errors
        df = pd.DataFrame(np.random.randint(0,100,size=(100, 8)), columns=FEATURES + [TARGET])

    # Preprocessing & Model Pipeline
    X = df[FEATURES]
    y = df[TARGET]
    X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, random_state=1)

    # Random Forest optimized with 200 trees for better generalization
    rf_model = RandomForestRegressor(n_estimators=200, random_state=1)
    rf_model.fit(X_train, y_train)
    
    return rf_model, X_valid, y_valid

model, X_valid, y_valid = load_and_train()
```

## 2. Model Evaluation and Insight Analysis
Once the model is trained, we evaluate its accuracy using Mean Absolute Error (MAE) and visualize how the predictions compare to actual values.

```python
# Generate validation predictions
val_predictions = model.predict(X_valid)
mae = mean_absolute_error(y_valid, val_predictions)

print(f"--- Model Analytics ---")
print(f"Validation MAE: ${mae:,.2f}")

# Portfolio-ready visualization: Top 5 Comparison
results_comparison = pd.DataFrame({
    'Actual Price': y_valid.head(),
    'Predicted Price': val_predictions[:5].round(2),
    'Difference': (y_valid.head() - val_predictions[:5]).round(2)
})

print("\nSample Predictions vs Actuals:")
display(results_comparison)
```
